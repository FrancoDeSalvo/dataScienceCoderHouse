```python
ws = wb['2020.02']
lista = [r[48].value for r in ws.iter_rows()]
lista
```

```python
#lista.pop(0)
lista.remove('Salario mensual NETO (en tu moneda local)')
lista[:10]
```

```python
data = list(filter(None, lista))
data[:200]
```

```python
for x in data:
    print(type(x))
```

```python
import datetime
cleanData = []
for d in data:
    if(type(d) == datetime.datetime):
        print("problemas con este dato:", d.timestamp())
    elif(type(d) == str):
         cleanData.append(float(d))
    else :
        cleanData.append(d)  
```

```python
for x in cleanData:
    print(type(x))
```

```python
print("Media: ", '{:0.2f}'.format(np.mean(cleanData)), "\nMediana:",'{:0.2f}'.format(np.median(cleanData)))
```
