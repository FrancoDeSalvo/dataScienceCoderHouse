```python
olimpicos.set_index("Year", inplace=True)
olimpicos.head()
```

```python
names = olimpicos['Athlete'].str.split(",", expand=True)
names.columns = ['Athlete_last_name', 'Athlete_name', 'delete' ]
names.head()
```

```python
names = names.drop(['delete'], axis=1)
names.head()
```

```python
olimpicos = pd.concat([olimpicos, names], axis=1)
olimpicos.head()
```

```python
olimpicos = olimpicos.drop(['Athlete'], axis=1)
olimpicos.head()
```
