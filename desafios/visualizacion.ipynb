```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_hr = pd.read_csv("HRDataset_v14.csv") 
```

```python
df_hr.head()
```

```python
df_hr.info()
```

```python
df_hr.describe().T
```

```python
fig, ax = plt.subplots(figsize=(14, 7))
ax.hist(df_hr["Salary"], bins=8, color="orange")
ax.set_title('Histograma de salarios')
ax.set_xlabel('Salarios USD')
ax.set_ylabel('Frecuencia')
```

```python
df_hr["GenderID"].values
```

```python
df_hr["GenderID"] = df_hr["GenderID"].astype(str)
df_hr.info()
```

```python
df_hr["GenderID"] = df_hr["GenderID"].replace(['0','1'],['female','male'])
df_hr.head()
```

```python
ax = sns.catplot(data=df_hr, kind='violin', x='GenderID', y ="Salary", hue='GenderID', split=True)
ax.set(xlabel='Genero', ylabel='Salario', title='Salarios por genero')
```

```python
df_hr["DateofHire"]= pd.to_datetime(df_hr['DateofHire'])
df_hr.info()        
```

```python
df_hr.set_index('DateofHire', inplace=True)
df_hr.head()
```

```python
df_hr.groupby(df_hr.index.year)['EmpID'].count().plot.bar()
```
