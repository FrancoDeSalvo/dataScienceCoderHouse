#### Calcular la proporción que hay entre personas con salarios >50k respecto del total para Entre personas de distinto sexo, Raza y Educacion. Por ejemplo cuál es el porcentaje de mujeres con un salario mayor a 50K y compararlo con el de los hombres. Lo mismo para personas de distintas razas y Educación


```python
groupedby_Sex = df_merge.groupby(['Sex','Target'])['ID'].count()
```

```python
groupedby_Sex.groupby(level=0).apply(lambda x: round((100 * x / x.sum()),2))
```

```python
groupedby_Sex.groupby(level=0).apply(lambda x: round((100 * x / len(df_merge)),2))
```

```python
groupedby_Race = df_merge.groupby(['Race','Target'])['ID'].count()
groupedby_Race.groupby(level=0).apply(lambda x: round((100 * x / x.sum()),2))
```

```python
groupedby_Education = df_merge.groupby(['Education','Target'])['ID'].count()
groupedby_Education.groupby(level=0).apply(lambda x: round((100 * x / x.sum()),2))
```
